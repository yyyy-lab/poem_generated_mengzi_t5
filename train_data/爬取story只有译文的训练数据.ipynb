```python


import requests
from bs4 import BeautifulSoup
import re
import requests

url = requests.get("https://so.gushiwen.cn/shiwens/default.aspx?tstr=%e5%86%ac%e5%a4%a9")#唐诗300

data = url.text
data2 = re.findall(r'<span><a href="(.*?)" target="_blank">(.*?)</a>(.*?)</span>', data)
# print(data2[0])

data = []
for i in data2:
    # print("-------------------------------------------")
    # print(str(i[1]))
    # print(str(i[2])[1:-1])

    url2 = 'https://so.gushiwen.cn/' + i[0]
    # url2 = i[0]
    # print(url2)
    # 发送HTTP请求
    try:
        response = requests.get(url2)
        # 处理正常的响应数据
    except requests.RequestException as e:
        print(f"请求发生错误")
        continue
        # 执行其他逻辑，或者直接使用 'continue' 跳过当前循环迭代

    # gushi = re.sub(r"[\[\]']", '',str(re.findall('name="description" content="(.*?)" /',response.text)))
    gushi = str(re.findall('name="description" content="(.*?)" /',response.text))[2:-2]
    gushi2 = re.sub(r'（[^（）]*）', '',  gushi)
    gushi1 = re.sub(r'\([^()]*\)', '', gushi2)
    # print(gushi)
    # 解析HTML页面
    soup = BeautifulSoup(response.content, 'html.parser')

    # 提取所需内容
    content = soup.get_text()

    # 打印内容
    # print(content)
    if content.rfind("意译") != -1:
      start_phrase = "意译"
    elif content.rfind("韵译") != -1:
      start_phrase = "韵译"
    else:
      start_phrase = "译文"

    end_phrase = "注释"

    start_index = content.rfind(start_phrase)
    # end_index = content.rfind(end_phrase, start_index)
    # 在 start_index 之后查找第一个 end_phrase 的位置
    end_index = content.find(end_phrase, start_index + len(start_phrase))

    if start_index == -1 or end_index == -1:
        print("未找到yiwen")
        continue
    else:
      start_index += len(start_phrase)
      extracted_content = content[start_index:end_index].strip()
      # print(extracted_content)

    print(f'诗名：{str(i[1])}，作者：{str(i[2])[1:-1]}，古诗内容：{gushi1},译文：{extracted_content}')
    data.append({"poem_name":str(i[1]),"author":str(i[2])[1:-1],"content":gushi1,"story":extracted_content})
```

```python
import json



# 指定要保存的JSON文件路径
file_path = '14poemdata.json'

# 将数据写入JSON文件
with open(file_path, 'w') as json_file:
    json.dump(data, json_file)
```

```python
def Classify_chinese_poem(poem):
        # 按逗号或句号拆分诗句
        lines = re.split(r'[，。]', poem)
        # 去掉空行
        lines = [line.strip() for line in lines if line.strip()]
        # 获取每行的字数
        line_lengths = [len(line) for line in lines]

        if len(lines) == 4:
            if all(length == 5 for length in line_lengths):
                return "五言绝句"
            elif all(length == 7 for length in line_lengths):
                return "七言绝句"
            else:
                return "诗"
        elif len(lines) == 8:
            if all(length == 5 for length in line_lengths):
                return "五言律诗"
            elif all(length == 7 for length in line_lengths):
                return "七言律诗"
            else:
                return "诗"
        else:
            return "诗"

print(Classify_chinese_poem("小雨丝丝欲网春，落花狼藉近黄昏。车尘不到张罗地，宿鸟声中自掩门。"))
```
