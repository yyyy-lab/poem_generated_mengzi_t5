```python
df_list = []

for i in range(50):
  url = "https://raw.githubusercontent.com/chinese-poetry/chinese-poetry/master/%E5%85%A8%E5%94%90%E8%AF%97/poet.tang." + str(i*1000) + ".json"
  df_list.append(pd.read_json(url))
```

```python
df = pd.concat(df_list)
df
```

```python
df['concat_paragraphs'] = [''.join(map(str, l)) for l in df['paragraphs']]
df = df[['author', 'title', 'concat_paragraphs']]
df
```

```python
import re

# 将content列中的字符串转换为数字并赋值给新列clasi
def Classify_chinese_poem(poem):
# 按逗号或句号拆分诗句
        lines = re.split(r'[，。]', poem)
        # 去掉空行
        lines = [line.strip() for line in lines if line.strip()]
        line_lengths = [len(line) for line in lines]

        if len(lines) == 4:
            if all(length == 5 for length in line_lengths):
                return "五言绝句"
            elif all(length == 7 for length in line_lengths):
                return "七言绝句"
            else:
                return "诗"
        elif len(lines) == 8:
            if all(length == 5 for length in line_lengths):
                return "五言律诗"
            elif all(length == 7 for length in line_lengths):
                return "七言律诗"
            else:
                return "诗"
        else:
            return "诗"

# print(Classify_chinese_poem("秦川雄帝宅，函谷壯皇居。綺殿千尋起，離宮百雉餘。連甍遙接漢，飛觀迥凌虛。雲日隱層闕，風煙出綺疎。"))
df['classification'] = df['concat_paragraphs'].apply(Classify_chinese_poem)
df
```

```python
filtered_df2 = df[df['classification'].isin(['五言绝句', '七言绝句'])]
```

```python
filtered_df1 = df[df['classification'].isin(['五言律诗', '七言律诗'])]
filtered_df1
```

```python
def convert_schinese(tchinese):
  return chinese_converter.to_simplified(tchinese)

filtered_df1['s_content'] = filtered_df1.apply(lambda row: convert_schinese(''.join(row.concat_paragraphs)), axis=1)
filtered_df1['s_title'] = filtered_df1.apply(lambda row: convert_schinese(''.join(row.title)), axis=1)
filtered_df1['s_author'] = filtered_df1.apply(lambda row: convert_schinese(''.join(row.author)), axis=1)

filtered_df2['s_content'] = filtered_df2.apply(lambda row: convert_schinese(''.join(row.concat_paragraphs)), axis=1)
filtered_df2['s_title'] = filtered_df2.apply(lambda row: convert_schinese(''.join(row.title)), axis=1)
filtered_df2['s_author'] = filtered_df2.apply(lambda row: convert_schinese(''.join(row.author)), axis=1)




my_df1 = filtered_df1
my_df2 = filtered_df2
my_df2
```

```python
my_df1 = my_df1[['classification', 's_content', 's_title','s_author']]
my_df1
```
